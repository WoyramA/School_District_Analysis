```python
# Dependencies
import pandas as pd
import os
```

```python
# Create DataFrame called student_df

new_full_student_data = os.path.join('../Resources/new_full_student_data.csv')
student_df = pd.read_csv(new_full_student_data)
```

```python
# Remove NaN
student_df = student_df.dropna(0)
student_df
```

```python
# Remove duplicats
student_df = student_df.drop_duplicates()
student_df
```

```python
student_df.loc[:,"grade"] = student_df.loc[:,"grade"].str.replace("th","")
student_df
```

```python
student_df.loc[:,"grade"] = student_df.loc[:,"grade"].astype("int")
student_df
```

```python
# Store row with minimum overall reading score in min_reading_row by using loc and the min_reading_score variable from Deliverable 3
min_reading_score = student_df["reading_score"].min()
min_reading_score_row = student_df.loc[student_df["reading_score"]==min_reading_score]
min_reading_score_row
```

```python
# Select all reading scores from the 10th graders at Dixon High School using loc with conditionals
all_reading_scores_10th_grade = student_df.loc[student_df.grade==10]
all_reading_scores_10th_grade
```

```python
# Find the mean reading score for all the student in Grades 11 and 12 combined by using conditional statements and loc or iloc
mean_reading_score_11thand12th_grades = student_df.loc[student_df.grade>10].mean
mean_reading_score_11thand12th_grades()
```
